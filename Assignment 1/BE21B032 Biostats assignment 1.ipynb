```python
data = data_be32.to_numpy()
data.shape
```

```python
m1 = np.mean(data)
m2 = moment(data - m1, moment=2)
m3 = moment(data - m1, moment=3)
m4 = moment(data - m1, moment=4)

print(f"First moment (sample mean): {1}")
print(f"Second moment: {m2}")
print(f"Third moment: {m3}")
print(f"Fourth moment: {m4}")
```

```python
beta_1 = np.power(m3,2) / np.power(m2,3) #skewness
beta_2 = m4 / np.power(m2,2)   #kurtosis
classification_term = m3 / (m2 ** 1.5)
print(beta_1)
print(beta_2)
print(f"Classification term: {classification_term}")
```

```python
data_be32_q2 = pd.read_csv('/content/drive/MyDrive/Important Folders & Files/BT2022/data_part2.csv')
data_be32_q2.info()
data_be32_q2.head()
```

```python
data_q2 = data_be32_q2.to_numpy()
print(data_q2)
data_be32_q2 = np.random.rand(120,10)
for i in range(120):
    for j in range(10):
        data_be32_q2[i][j] = data[data_q2[i][j]]
print(data_be32_q2)               
```

```python
z = np.sum(data_be32_q2,axis = 1)/np.sqrt(10)
print(z)
```

```python
mz = np.mean(z)
sz = np.std(z)
alpha = 0.0180062
l = len(z)
n = math.sqrt(l)

t = sp.t.ppf(q = alpha/2, df = l - 1)
sigma = sz/n
error_margin = sigma*abs(t)
interval = ((mz - error_margin)/np.sqrt(10), (mz + error_margin)/np.sqrt(10))
print(interval)
```

```python
chi1 = sp.chi2.ppf(q = alpha/2, df = l - 1)
chi2 = sp.chi2.ppf(q = 1 - (alpha/2), df = l - 1)
nu = len(z)-1
lower = (nu*np.var(z))/chi2
upper = (nu*np.var(z))/chi1
print((lower, upper))
```
